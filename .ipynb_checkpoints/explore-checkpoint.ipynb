# Polymarket BTC 15-Min Data Explorer

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load all recorded windows
df = pd.read_parquet("data/")
df["ts"] = pd.to_datetime(df["ts_ms"], unit="ms", utc=True)
print(f"Rows: {len(df):,}  |  Columns: {len(df.columns)}  |  Windows: {df['market_slug'].nunique()}")
df.head()
```

## Chainlink BTC Price

```python
fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(df["ts"], df["chainlink_btc"], linewidth=0.8)
ax.set_ylabel("BTC/USD")
ax.set_title("Chainlink BTC/USD Over Time")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Best Bid / Ask (Up outcome)

```python
fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(df["ts"], df["best_bid_up"], label="Bid (Up)", alpha=0.7)
ax.plot(df["ts"], df["best_ask_up"], label="Ask (Up)", alpha=0.7)
ax.fill_between(df["ts"], df["best_bid_up"], df["best_ask_up"], alpha=0.15)
ax.set_ylabel("Price")
ax.set_title("Up Outcome: Best Bid / Ask")
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Mid Price & Spread

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 6), sharex=True)

ax1.plot(df["ts"], df["mid_up"], label="Mid (Up)", linewidth=0.8)
ax1.plot(df["ts"], df["mid_down"], label="Mid (Down)", linewidth=0.8)
ax1.set_ylabel("Mid Price")
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2.plot(df["ts"], df["spread_up"], label="Spread (Up)", linewidth=0.8)
ax2.plot(df["ts"], df["spread_down"], label="Spread (Down)", linewidth=0.8)
ax2.set_ylabel("Spread")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Book Imbalance (Top 5 Levels)

```python
fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(df["ts"], df["imbalance5_up"], label="Imbalance (Up)", linewidth=0.8)
ax.plot(df["ts"], df["imbalance5_down"], label="Imbalance (Down)", linewidth=0.8)
ax.axhline(0, color="gray", linestyle="--", alpha=0.5)
ax.set_ylabel("Imbalance (bid-ask)/(bid+ask)")
ax.set_title("Order Book Imbalance")
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Single Window Deep Dive

```python
# Pick the latest window
latest = df[df["market_slug"] == df["market_slug"].iloc[-1]].copy()
print(f"Window: {latest['market_slug'].iloc[0]}")
print(f"Rows: {len(latest)}  |  Duration: {latest['time_remaining_s'].max():.0f}s")

fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

axes[0].plot(latest["ts"], latest["chainlink_btc"], linewidth=0.8)
if latest["window_start_price"].notna().any():
    axes[0].axhline(latest["window_start_price"].dropna().iloc[0], color="red", linestyle="--", label="Start Price")
axes[0].set_ylabel("BTC/USD")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(latest["ts"], latest["best_ask_up"], label="P(Up) ask", linewidth=0.8)
axes[1].plot(latest["ts"], latest["best_ask_down"], label="P(Down) ask", linewidth=0.8)
axes[1].set_ylabel("Ask Price")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

axes[2].plot(latest["ts"], latest["time_remaining_s"], linewidth=0.8, color="green")
axes[2].set_ylabel("Time Remaining (s)")
axes[2].grid(True, alpha=0.3)

plt.suptitle(f"Window: {latest['market_slug'].iloc[0]}", fontsize=13)
plt.tight_layout()
plt.show()
```
